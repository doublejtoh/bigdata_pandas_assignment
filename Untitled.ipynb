```python
import pandas as pd
import numpy as np
df_air = pd.read_csv("airbnb.csv")
df_air
```

```python
df_air['overall_satisfaction'] = df_air.overall_satisfaction.fillna(3)
df_air
```

```python

df = df_air.set_index(['room_id', 'host_id'])
df['total_score'] = df['overall_satisfaction'] + df['reviews'] * 0.378

df_ascend = df.sort_values('total_score', ascending=True)
df_descend = df.sort_values('total_score', ascending=False)
df_ascend = df_ascend[['total_score']]
df_descend = df_descend[['total_score']]
df_ascend.to_csv('sorted_total_score_ascend.csv')
df_descend.to_csv('sorted_total_score_descend.csv')
```

```python

newdf = df_air.copy()

newdf_groupby_neighborhood = newdf.groupby('neighborhood')

def s(group):
    group['avg_of_reviews'] = group['reviews'].mean()
    group['avg_of_overall_satisfaction'] = group['overall_satisfaction'].mean()
    group['avg_of_price'] = group['price'].mean()
    group['max_of_reviews'] = group['reviews'].max()
    group['min_of_reviews'] = group['reviews'].min()
    group['max_of_price'] = group['price'].max()
    group['min_of_price'] = group['price'].min()
    return group

newdf = newdf_groupby_neighborhood.apply(s)
newdf = newdf.set_index('neighborhood')
newdf = newdf[['avg_of_reviews','avg_of_overall_satisfaction', 'avg_of_price', 'max_of_reviews', 'min_of_reviews', 'max_of_price', 'min_of_price']]
newdf_ascending = newdf.sort_index(axis=0, ascending=True)
newdf_ascending = newdf_ascending[~newdf_ascending.index.duplicated(keep='first')]
newdf_ascending.to_csv('sorted_neighborhood_factors.csv')
newdf_ascending
```

```python
newdf = df_air.copy()

newdf_groupby_neighborhood = newdf.groupby(pd.cut(newdf['price'], [0, 100, 200, 300, 400, 500, 1000, 5000]))


price_list = ['0-100', '100-200', '200-300', '300-400', '400-500', '500-1000', '1000-5000']
accommodates_average_list = []
accommodates_median_list = []
bedrooms_average_list = []
bedrooms_median_list = []
reviews_average_list = []
reviews_median_list = []
neighbor_string_list = []
neighbor_length_list = []
for idx, data in newdf_groupby_neighborhood:
    accommodates_average_list.append(data['accommodates'].mean())
    accommodates_median_list.append(data['accommodates'].median())
    bedrooms_average_list.append(data['bedrooms'].mean())
    bedrooms_median_list.append(data['bedrooms'].median())
    reviews_average_list.append(data['reviews'].mean())
    reviews_median_list.append(data['reviews'].median())
    neighbor_string_list.append(data['neighborhood'].str.cat(sep=','))
    neighbor_length_list.append(len(data['neighborhood']))

final_df_config_json = {
    'accommodates_average' : accommodates_average_list,
    'accommodates_median' : accommodates_median_list,
    'bedrooms_average'     : bedrooms_average_list,
    'bedrooms_median'     : bedrooms_median_list,
    'reviews_average'      : reviews_average_list,
    'reviews_median'      : reviews_median_list,
    'neighbor_list'        : neighbor_string_list,
    'length'               : neighbor_length_list,
    'price'                : price_list
}
final_df = pd.DataFrame(final_df_config_json)
final_df = final_df.set_index(['price'])
final_df
    
    

# def s(group):
#     group['accommodates_average'] = group['accommodates'].mean()
#     return group

# newdf = newdf_groupby_neighborhood.apply(s)
# newdf
# newdf_groupby_neighborhood
```
